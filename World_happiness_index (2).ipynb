# World happiness report 2021

```python
data= pd.read_csv('world-happiness-report-2021.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.apply(lambda x: len(x.unique()))
```

```python
top10 = data.head(10)
bottom10=data.tail(10)
```

#Happiest countries in 2021

### **Top 10** 

```python
plt.figure(1, figsize=(10,8))
sns.barplot(x='Ladder score', y='Country name', data= top10);
```

### **Bottom 10**

```python
plt.figure(1, figsize=(10,8))
sns.barplot(x='Ladder score', y='Country name', data= bottom10);
```

# Ladder score vs GDP

```python
plt.figure(1,figsize=(22,15))
sns.regplot(x='Ladder score',y='Logged GDP per capita',data= data, scatter_kws={"s": 100})
for i in range(len(data)):
    plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10)
plt.show()
```

* **There is a linear relationship between GDP and Ladder score.**
